```python
register = {"A":0,
            "B":0,
            "C":0}

with open("inputs.txt", "r") as f:
    for line in f:
        if len(line.strip().split(":")) > 1:
            key, value = line.strip().split(":")
        else: continue
        if "A" in key:
            register["A"] = int(value)
        elif "B" in key:
            register["B"] = int(value)
        elif "C" in key:
            register["C"] = int(value)
        
        elif "Pro" in key:
            program = list(map(int, value.split(",")))
```

```python
def runInstructionRun(program):

    i = 0
    output = []
    programPointer = 0
    while programPointer < len(program) - 1:

        currentProgram = program[programPointer]
        currentOperand = program[programPointer+1]

        # print("current operations and param, ", (currentProgram,currentOperand))

        match currentProgram:

            case 0:
                adv(currentOperand)
                programPointer += 2
            case 1:
                bxl(currentOperand)
                programPointer += 2
            case 2:
                bst(currentOperand)
                programPointer += 2

            case 3:
                tempVal = jnz(currentOperand)
                if tempVal == "cont":
                    programPointer += 2
                else:
                    programPointer = tempVal
            case 4:
                bxc(currentOperand)
                programPointer += 2

            case 5:
                output.append(out(currentOperand))
                programPointer += 2
            case 6:
                bdv(currentOperand)
                programPointer += 2
            case 7:
                cdv(currentOperand)
                programPointer += 2
    return output
```

```python
output = runInstructionRun(program)
print(','.join(str(x) for x in output))
```

```python
# collection of functions to perform at each step
# op 0
def adv( operand):
    num = register["A"]
    if operand in {0, 1, 2, 3}:
        result = num//(2**operand)
    elif operand == 4:
        result = num//(2**register["A"])
    elif operand == 5:
        result = num//(2**register["B"])
    elif operand == 6:
        result = num//(2**register["C"])
    else:
        raise ValueError("Invaliud Program")
    register["A"] = result

# op 1
def bxl(operand):
    value = register["B"]
    register["B"] = value^operand

# op 2
def bst(operand):
    if operand in {0, 1, 2, 3}:
        result = operand % 8
    elif operand == 4:
        result = register["A"] % 8
    elif operand == 5:
        result = register["B"] % 8
    elif operand == 6:
        result = register["C"] % 8
    else:
        raise ValueError("Invaliud Program")
    register["B"] = result

# op 3
def jnz(operand):
    if register["A"] != 0:
        return operand
    else: 
        return  "cont" 
    
# op 4
def bxc(operand):
    value1 = register["B"]
    value2 = register["C"]
    register["B"] = value1 ^ value2

# op 5
def out(operand):
    if operand in {0, 1, 2, 3}:
        result = operand % 8
    elif operand == 4:
        result = register["A"] % 8
    elif operand == 5:
        result = register["B"] % 8
    elif operand == 6:
        result = register["C"] % 8
    else:
        raise ValueError("Invaliud Program")
    return result

#op 6
def bdv(operand):
    num = register["A"]
    if operand in {0, 1, 2, 3}:
        result = num//2**operand
    elif operand == 4:
        result = num//2**register["A"]
    elif operand == 5:
        result = num//2**register["B"]
    elif operand == 6:
        result = num//2**register["C"]
    else:
        raise ValueError("Invaliud Program")
    register["B"] = result


# op7
def cdv(operand):
    num = register["A"]
    if operand in {0, 1, 2, 3}:
        result = num//2**operand
    elif operand == 4:
        result = num//2**register["A"]
    elif operand == 5:
        result = num//2**register["B"]
    elif operand == 6:
        result = num//2**register["C"]
    else:
        raise ValueError("Invaliud Program")
    register["C"] = result
```
